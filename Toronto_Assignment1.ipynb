# Segmenting and Clustering Neighborhoods in Toronto (Assignment 2)

## This is scraping part

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

print("Import done!")
```

```python
#add url and request
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
web_text =  requests.get(url).text
```

```python
soup = BeautifulSoup(web_text, 'html.parser')
#print(soup.prettify())
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
```

```python
#s = My_table.findAll(text=True)
#My_table.get_text(strip=True)
```

```python
text = My_table.get_text()
d = text.split('\n\n\n')
data = []
d[0:5]
```

```python
#d[0] = d[0].split('\n')
```

```python
#d[0]
```

```python
for index,r in enumerate(d):
    row = r.split('\n')
    data.append(row)
```

```python
data[0:5]
```

```python
del data[0][0:2]
data
```

## Convert scraping to dataframe

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df.head()
```

```python
df.drop(df.columns[4:], axis=1, inplace=True) #drop None columns
```

```python
df.head()
```

```python
df.replace("Not assigned", np.nan, inplace = True) 
df.head(5)
```

```python
df1 = df[:]
```

```python
df1.head()
```

```python
df1.dropna(axis=0, inplace=True)
df1.reset_index(drop=True, inplace=True)
df1.rename(columns={'Postcode':'PostalCode'}, inplace=True)
df1.rename(columns={'Neighbourhood':'Neighborhood'}, inplace=True)
df1.head()
```

```python
df1 = df1.groupby(['PostalCode','Borough'])['Neighborhood'].apply(', '.join).reset_index()
```
